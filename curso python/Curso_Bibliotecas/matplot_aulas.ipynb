## Matplotlib


```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
dados_x = np.linspace(0,2*np.pi,500)
dados_y = np.sin(dados_x)

x_2 = np.linspace(0,2*np.pi,500)
y_2 = np.sin(x_2+np.pi/2)
```

```python
plt.plot(dados_x,dados_y) #plotar uma curva ou um gráfico
plt.plot(x_2,y_2) #plotar um outro gráfico junto com o primeiro
plt.xlabel('x') # legenda do eixo x
plt.ylabel('y') #legenda do eixo y
plt.grid() #linhas do gráfico 
```

```python
#Segunda maneira de se plotar

fig, ax = plt.subplots()
ax.plot(dados_x,dados_y,'b-',label ='func_a') #legenda do gráfico
ax.grid()
ax.set (xlabel= 'x', ylabel = 'y',title = 'seno') #maneira mais prática de se determinra os nomes doseixos
ax.legend()
```
